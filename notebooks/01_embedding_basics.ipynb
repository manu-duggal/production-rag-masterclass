## Hypothesis

Before running any experiments:

- Similar sentences should produce similar embeddings.
- Unrelated sentences should produce dissimilar embeddings.
- Every embedding produced by the same model should have the same number of dimensions.
- Individual values inside an embedding are not directly interpretable.

## Step 5 — Compare Similar Sentences

```python
from sentence_transformers import util
```

```python
sentence1 = "The doctor examined the patient."
sentence2 = "The physician checked the patient."
sentence3 = "The banana fell from the tree."

embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)
embedding3 = model.encode(sentence3)

similarity_12 = util.cos_sim(embedding1, embedding2)
similarity_13 = util.cos_sim(embedding1, embedding3)

print(similarity_12)
print(similarity_13)
```

```python
similarity_11 = util.cos_sim(embedding1, embedding1)
print(similarity_11)
```

```python
sentence3 = "The surgeon performed the operation."
embedding3 = model.encode(sentence3)
similarity_13 = util.cos_sim(embedding1, embedding3)
print(similarity_13.item())
```

```python
sentence3 = "The mechanic repaired the engine."
embedding3 = model.encode(sentence3)
similarity_13 = util.cos_sim(embedding1, embedding3)
print(similarity_13.item())
```

```python
sentence3 = "The patient was examined by the doctor."
embedding3 = model.encode(sentence3)
similarity_13 = util.cos_sim(embedding1, embedding3)
print(similarity_13.item())
```

```python
sentence3 = "The patient examined the doctor."
embedding3 = model.encode(sentence3)
similarity_13 = util.cos_sim(embedding1, embedding3)
print(similarity_13.item())
```

```python
sentence1 = "Refunds are allowed within 30 days."
sentence2 = "Refunds are not allowed within 30 days."

embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

similarity_12 = util.cos_sim(embedding1, embedding2)
print(similarity_12.item()) 
```
